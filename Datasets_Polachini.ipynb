# Descrição das Variáveis Potencialmente Interessantes

## 1. Saúde Pública (`health_data.csv`)

**Contexto:**  
Este conjunto de dados busca identificar fatores de risco e padrões relacionados a doenças e ao uso do sistema de saúde.

**Variáveis relevantes:**  
- **Age (Idade):** Importante para segmentar grupos etários e identificar populações vulneráveis.  
- **BMI (Índice de Massa Corporal):** Auxilia na avaliação do estado nutricional e risco de doenças crônicas.  
- **Smoker (Fumante):** Hábito que impacta diretamente no risco de doenças cardiovasculares e respiratórias.  
- **Chronic_Disease (Doença Crônica):** Indica condições de saúde que influenciam no cuidado médico e prognóstico.  
- **Diagnosis (Diagnóstico):** Permite análise das doenças mais prevalentes.  
- **Hospital_Visits (Visitas Hospitalares):** Reflete o uso dos serviços de saúde e gravidade do quadro.  
- **Income_Level (Renda):** Relação entre condições socioeconômicas e saúde.  
- **Region (Região):** Ajuda a identificar disparidades geográficas no acesso e qualidade do atendimento.

---

## 2. Insegurança Alimentar (`food_insecurity_data.csv`)

**Contexto:**  
Avaliar as condições de insegurança alimentar em famílias para subsidiar políticas públicas e ações sociais.

**Variáveis relevantes:**  
- **Food_Shortage_Days (Dias de Falta de Alimentos):** Indicador direto da insegurança alimentar.  
- **Child_Malnutrition (Desnutrição Infantil):** Impacto crítico na saúde e desenvolvimento infantil.  
- **Income_Level (Renda):** Determinante central da capacidade de aquisição alimentar.  
- **Access_Markets (Acesso a Mercados):** Facilita o consumo de alimentos frescos e nutritivos.  
- **Govt_Assistance (Assistência Governamental):** Avalia o suporte público recebido.  
- **Food_Diversity_Score (Diversidade Alimentar):** Mede qualidade e variedade da alimentação.  
- **Household_Size (Tamanho da Família):** Influencia na demanda por alimentos.

---

## 3. Corredores Ecológicos (`corridors_data.csv`)

**Contexto:**  
Analisar a estrutura, conectividade e ameaças aos corredores ecológicos, essenciais para a conservação da biodiversidade.

**Variáveis relevantes:**  
- **Length_km (Comprimento):** Influencia a funcionalidade dos corredores para o deslocamento da fauna.  
- **Width_km (Largura):** Importante para abrigo e qualidade do habitat.  
- **Connectivity_Score (Índice de Conectividade):** Avalia a ligação entre habitats isolados.  
- **Species_Richness (Riqueza de Espécies):** Indicador da biodiversidade local.  
- **Endangered_Species (Espécies Ameaçadas):** Prioridade para ações de conservação.  
- **Human_Pressure_Index (Pressão Humana):** Avalia impactos antrópicos negativos.  
- **Illegal_Activity_Reports (Relatórios de Atividades Ilegais):** Indica ameaças reais e urgência de intervenções.  
- **Protection_Level (Nível de Proteção):** Grau de proteção legal ou conservacionista.  
- **Restoration_Projects (Projetos de Restauração):** Presença de iniciativas para recuperação ambiental.
