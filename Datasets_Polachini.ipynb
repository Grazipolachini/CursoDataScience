# Descrio das Variveis Potencialmente Interessantes

## 1. Sade Pblica (`health_data.csv`)

**Contexto:**  
Este conjunto de dados busca identificar fatores de risco e padres relacionados a doenas e ao uso do sistema de sade.

**Variveis relevantes:**  
- **Age (Idade):** Importante para segmentar grupos etrios e identificar populaes vulnerveis.  
- **BMI (ndice de Massa Corporal):** Auxilia na avaliao do estado nutricional e risco de doenas crnicas.  
- **Smoker (Fumante):** Hbito que impacta diretamente no risco de doenas cardiovasculares e respiratrias.  
- **Chronic_Disease (Doena Crnica):** Indica condies de sade que influenciam no cuidado mdico e prognstico.  
- **Diagnosis (Diagnstico):** Permite anlise das doenas mais prevalentes.  
- **Hospital_Visits (Visitas Hospitalares):** Reflete o uso dos servios de sade e gravidade do quadro.  
- **Income_Level (Renda):** Relao entre condies socioeconmicas e sade.  
- **Region (Regio):** Ajuda a identificar disparidades geogrficas no acesso e qualidade do atendimento.

## 2. Insegurana Alimentar (`food_insecurity_data.csv`)

**Contexto:**  
Avaliar as condies de insegurana alimentar em famlias para subsidiar polticas pblicas e aes sociais.

**Variveis relevantes:**  
- **Food_Shortage_Days (Dias de Falta de Alimentos):** Indicador direto da insegurana alimentar.  
- **Child_Malnutrition (Desnutrio Infantil):** Impacto crtico na sade e desenvolvimento infantil.  
- **Income_Level (Renda):** Determinante central da capacidade de aquisio alimentar.  
- **Access_Markets (Acesso a Mercados):** Facilita o consumo de alimentos frescos e nutritivos.  
- **Govt_Assistance (Assistncia Governamental):** Avalia o suporte pblico recebido.  
- **Food_Diversity_Score (Diversidade Alimentar):** Mede qualidade e variedade da alimentao.  
- **Household_Size (Tamanho da Famlia):** Influencia na demanda por alimentos.

---

## 3. Corredores Ecolgicos (`corridors_data.csv`)

**Contexto:**  
Analisar a estrutura, conectividade e ameaas aos corredores ecolgicos, essenciais para a conservao da biodiversidade.

**Variveis relevantes:**  
- **Length_km (Comprimento):** Influencia a funcionalidade dos corredores para o deslocamento da fauna.  
- **Width_km (Largura):** Importante para abrigo e qualidade do habitat.  
- **Connectivity_Score (ndice de Conectividade):** Avalia a ligao entre habitats isolados.  
- **Species_Richness (Riqueza de Espcies):** Indicador da biodiversidade local.  
- **Endangered_Species (Espcies Ameaadas):** Prioridade para aes de conservao.  
- **Human_Pressure_Index (Presso Humana):** Avalia impactos antrpicos negativos.  
- **Illegal_Activity_Reports (Relatrios de Atividades Ilegais):** Indica ameaas reais e urgncia de intervenes.  
- **Protection_Level (Nvel de Proteo):** Grau de proteo legal ou conservacionista.  
- **Restoration_Projects (Projetos de Restaurao):** Presena de iniciativas para recuperao ambiental.
